```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("genre_original.csv")
df.head(10)
```

```python
label_encoder = LabelEncoder()

df['Value'] = label_encoder.fit_transform(df['Value'])

label_encoding_table = pd.DataFrame({
    'Value': label_encoder.classes_,
    'encoding': label_encoder.transform(label_encoder.classes_)
})
```

```python
df
```

```python
df.to_csv('genre.csv', index=False)
```
